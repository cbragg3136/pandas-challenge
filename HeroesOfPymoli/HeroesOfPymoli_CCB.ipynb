```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data_file = "../../Instructions/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data_file)
purchase_data_df.head()
```

```python
#Display the total number of players
total_players = len(purchase_data_df['SN'].unique())
total_players_df = pd.DataFrame({'Total Players' : [total_players]})
total_players_df
```

```python
#Purchasing Analysis (Total) = Number of Unique Items, Average Purchase Price, Total Number of Purchases, Total Revenue
unique_items = len(purchase_data_df['Item ID'].unique())
avg_price = purchase_data_df['Price'].mean()
total_purchases = purchase_data_df['Purchase ID'].count()
revenue = purchase_data_df['Price'].sum()

purchasing_summary_df = pd.DataFrame({'Number of Unique Items' : [unique_items],
                                     'Average Purchase Price': '${0}'.format(round(avg_price,2)),
                                     'Number of Purchases': [total_purchases],
                                     'Total Revenue': '${0}'.format(round(revenue,2))})
purchasing_summary_df
```

```python
#gender demographics = Percentage and Count of Male, Female, and other/non-disclosed Players
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html
#https://stackoverflow.com/questions/35661968/add-a-percent-sign-to-a-dataframe-column-in-python

gender_sn_df = purchase_data_df[['Gender', 'SN']]
renamed_gender_sn_df = gender_sn_df.rename(columns = {'SN': 'Total Count'})
unique_sn_df = renamed_gender_sn_df.drop_duplicates(subset=['Total Count'])
group_df = unique_sn_df.groupby(['Gender'])
total_count_df = group_df.count()

percentage = total_count_df['Total Count']/total_count_df['Total Count'].sum() * 100
total_count_df['Percentage of Players'] = round(percentage,2).astype(str) + '%'
total_count_df
```

```python
# Purchasing Analysis (Gender) - purchase count, avg. purchase price, avg. purchase total per person etc. by gender
purch_count = purchase_data_df.groupby(['Gender']).count()['Purchase ID']
avg_price_group = purchase_data_df.groupby(['Gender']).mean()['Price']
total_value = purchase_data_df.groupby(['Gender']).sum()['Price']
purch_per_person = total_value/total_count_df['Total Count']

purch_analysis_gender_df = pd.DataFrame({'Purchase Count' : purch_count,
                                        'Average Purchase Price' : avg_price_group.map("${:,.2f}".format),
                                       'Total Purchase Value': total_value.map("${:,.2f}".format),
                                       'Avg Total Purchase per Person': purch_per_person.map("${:,.2f}".format)})
purch_analysis_gender_df
```

```python
# Age Demographics - Establish bins for ages, Categorize the existing players using the age bins. Hint: use pd.cut(), Calculate the numbers and percentages by age group
age_df = purchase_data_df[['Age', 'SN']]

unique_age_df = age_df.drop_duplicates(subset=['SN'])

bins = [1, 9, 14, 19, 24, 29, 34, 39, 50]
labels= ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
unique_age_df['Age Group'] = pd.cut(unique_age_df['Age'], bins, labels=labels)

age_groups = unique_age_df.groupby('Age Group')
count_age_groups = age_groups['Age'].count()

perc_players = age_groups['Age'].count()/age_groups['Age'].count().sum() * 100

age_demographics_df = pd.DataFrame({'Total Count': count_age_groups,
                                   'Percentage of Players': round(perc_players,2).astype(str) + '%'})
age_demographics_df
```

```python
# Purchasing Analysis (Age) - Bin the purchase_data df by age, Run to obtain purchase count, avg. purchase price, avg. purchase total per person

bins = [1, 9, 14, 19, 24, 29, 34, 39, 50]
labels= ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data_df['Age Group'] = pd.cut(purchase_data_df['Age'], bins, labels=labels)

purch_count_age = purchase_data_df.groupby(['Age Group']).count()['Purchase ID']
avg_price_age = purchase_data_df.groupby(['Age Group']).mean()['Price']
total_value_age = purchase_data_df.groupby(['Age Group']).sum()['Price']
purch_per_person = total_value_age/count_age_groups

purch_analysis_age_df = pd.DataFrame({'Purchase Count' : purch_count_age,
                                        'Average Purchase Price' : avg_price_age.map("${:,.2f}".format),
                                       'Total Purchase Value': total_value_age.map("${:,.2f}".format),
                                       'Avg Total Purchase per Person': purch_per_person.map("${:,.2f}".format)})
purch_analysis_age_df
```

```python
# Top Spenders = Sort the total purchase value column in descending order
#https://jakevdp.github.io/PythonDataScienceHandbook/03.08-aggregation-and-grouping.html
spenders_df =  purchase_data_df[['SN', 'Price', 'Purchase ID']]
spenders_count = spenders_df.groupby(['SN']).count()['Purchase ID']
spenders_sum = spenders_df.groupby(['SN']).sum()['Price']
avg_price_sn = spenders_sum / spenders_count
avg_price_sn

# # total_purch_value_df = spenders_grouped_df.sum().sort_values('Price', ascending=False)


top_spenders_df = pd.DataFrame({'Purchase Count': spenders_count,
                                'Average Purchase Price': avg_price_sn.map("${:,.2f}".format),
                                'Total Purchase Value' : spenders_sum,
                               })
total_purchase_value_df = top_spenders_df.sort_values('Total Purchase Value', ascending = False)
total_purchase_value_df.head(5)
```

```python
items_df = purchase_data_df[['Item ID', 'Item Name', 'Price']]

items_grouped = items_df.groupby(['Item ID', 'Item Name'])

item_count = items_grouped['Item Name'].count()
item_sum = items_grouped['Price'].sum()
avg_price_item = item_sum / item_count

most_popular_items_df = pd.DataFrame({'Purchase Count': item_count,
                                     'Item Price': avg_price_item.map("${:.2f}".format),
                                     'Total Purchase Value' : item_sum})
most_popular_df = most_popular_items_df.sort_values('Purchase Count', ascending = False)
most_popular_df.head(5)
```

```python
most_profitable_df = most_popular_items_df.sort_values('Total Purchase Value', ascending = False)
most_profitable_df.head(5)
```
